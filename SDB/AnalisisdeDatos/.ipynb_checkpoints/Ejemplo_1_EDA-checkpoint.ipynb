#### Carga de Datos desde GitHub

El siguiente código es el mismo que habíamos usado para descargar un archivo **`.csv`** desde una URL. Ahora, como el dataset que necesitamos, llamado **`tips`**, se encuentra alojado en GitHub, sólo debemos asegurarnos de usar el enlace **raw** para acceder directamente al archivo de datos.

```python
# Cargar dataset 'tips.csv' que viene con Seaborn
import pandas as pd

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips = pd.read_csv(url, sep=',')
```

El dataset **Tips (Propinas)** contiene los siguientes campos:

* **`total_bill`** (Cuenta Total): Representa el monto total de la factura pagada por los clientes, incluyendo la propina. Se refiere al costo total de la comida y cualquier otro servicio recibido en el restaurante.

* **`tip`** (Propina): Es el monto de dinero dejado por los clientes a los empleados del restaurante por el servicio recibido. Es una forma de agradecimiento por el servicio proporcionado.

* **`sex`** (Sexo): Indica el género del cliente que dejó la propina. Generalmente se clasifica como masculino o femenino.

* **`smoker`** (Fumador): Especifica si el grupo de clientes incluía fumadores. Este dato puede ser relevante en contextos donde los restaurantes tienen o no áreas designadas para fumadores, lo que podría influir en la experiencia del servicio.

* **`day`** (Día): Indica el día de la semana en el que los clientes visitaron el restaurante. Esto puede ser útil para analizar patrones de visitas durante la semana.

* **`time`** (Hora): Refleja si la visita al restaurante fue durante el almuerzo o la cena. Esta distinción puede ser relevante para entender las preferencias de los clientes y el volumen de negocio en diferentes momentos del día.

* **`size`** (Tamaño): Se refiere al número de personas en el grupo de clientes que visitó el restaurante. Este dato es importante para entender la relación entre el tamaño del grupo y la propina dejada, así como el consumo total.

#### Estadísticas Descriptivas

Las **variables numéricas** son aquellas que representan cantidades numéricas y pueden tomar valores enteros o decimales. Son fundamentales para la mayoría de los análisis estadísticos y matemáticos.

El método **describe()** en Pandas proporciona un resumen estadístico de las columnas numéricas en un DataFrame.

Este resumen incluye estadísticas descriptivas como:

* **`count`**: Número total de entradas no nulas.

* **`mean`**: Promedio de los valores.

* **`std`**: Desviación estándar, que mide la dispersión de los datos respecto al promedio.

* **`min`**: El valor mínimo.

* **`25%`** (Primer cuartil): El valor debajo del cual se encuentra el 25% de los datos.

* **`50%`** (Mediana): El valor medio que divide el conjunto de datos en dos partes iguales.

* **`75%`** (Tercer cuartil): El valor debajo del cual se encuentra el 75% de los datos.

* **`max`**: El valor máximo.

```python
# Variables Numéricas: Estadísticas Descriptivas
print(tips.describe())
```

Las **variables categóricas** son aquellas que contienen valores que representan categorías. Estas variables pueden ser **nominales** (sin un orden inherente) o **ordinales** (con un orden lógico).


El método **`value_counts()`** es utilizado para contar la frecuencia de cada categoría en una variable categórica. Es útil para entender la distribución de las categorías dentro de un conjunto de datos.


```python
# Variables Categóricas: Obtener frecuencias
print("Variable Categórica 'sex':\n", tips['sex'].value_counts())

print("\nVariable Categórica 'day':\n", tips['day'].value_counts())
```

#### Limpieza (Nulos)

**`isnull() `** es un método de pandas que identifica si cada celda en un DataFrame contiene un valor nulo (**`NaN`**). Retorna un DataFrame del mismo tamaño que el original, donde cada celda tiene un valor booleano indicando si es nula (**`True`**) o no (**`False`**).

Al encadenar el método **`sum()`** a **`isnull()`**, pandas sumará los valores **`True`** (considerados como **`1`**) para cada columna, proporcionando así un recuento *total de valores nulos por variable*. Esto es esencial para evaluar rápidamente cuántos datos faltantes contiene cada columna.

```python
# Identificar valores nulos
print(tips.isnull().sum())
```

Una forma directa de manejar los valores nulos es *eliminar las filas o columnas* que los contienen usando **`dropna()`**. Sin embargo, este enfoque no suele ser el ideal, ya que podría resultar en la pérdida de datos potencialmente valiosos.

#### Matriz de Correlación

La **Matriz de Correlación** es una tabla que muestra los coeficientes de correlación entre varias variables numéricas. Cada celda de la matriz muestra el coeficiente de correlación entre dos variables.

El **Coeficiente de Correlación**, típicamente el **Coeficiente de Correlación de Pearson**, mide la relación lineal entre dos variables numéricas.

Los valores del coeficiente pueden variar entre -1 y 1, donde:

* **`1`** indica una **correlación positiva** perfecta (a medida que una variable aumenta, la otra también lo hace en proporción directa).

* **`-1`** indica una **correlación negativa** perfecta (a medida que una variable aumenta, la otra disminuye en proporción directa).

* **`0`** indica que **no hay correlación** lineal entre las variables.

```python
# Matriz de Correlación para Variables Numéricas
# Crear un subconjunto con solo las variables numéricas
tips_numeric = tips.select_dtypes(include=['float64', 'int64'])

# Ahora, calcular y mostrar la matriz de correlación para este subconjunto
print(tips_numeric.corr())
```

#### Primeras Visualizaciones con Matplotlib

**`matplotlib`** es una biblioteca de visualización en Python que permite crear una amplia gama de gráficos y diagramas de manera flexible.

**`plt`** es el alias comúnmente usado para **`matplotlib.pyplot`**, que es el módulo de **`matplotlib`** diseñado para un estilo de uso procedimental, inspirado en MATLAB.

##### Histograma

El método **`.hist()`** se utiliza para generar un histograma, que es un tipo de gráfico que muestra la distribución de frecuencias de un conjunto de datos numéricos.

Cada barra del histograma representa la **frecuencia** (cantidad de veces) que los valores de datos caen dentro de un **rango** específico (bin o intervalo).

**`plt.title()`** permite añadir un título al gráfico.

**`plt.xlabel()`** añade una etiqueta al eje X del gráfico, brindando claridad sobre la variable que se está analizando.

**`plt.ylabel()`** añade una etiqueta al eje Y del gráfico.

**`plt.show()`** se llama al final para mostrar el gráfico generado.

* Histograma del Total de la Cuenta: **`total_bill`**

```python
# Histograma con Matplotlib
import matplotlib.pyplot as plt

# Histograma de la variable 'total_bill'
tips['total_bill'].hist()
plt.title('Distribución del Total de la Cuenta (total_bill)')
plt.xlabel('Total de la Cuenta (total_bill)')
plt.ylabel('Frecuencia')
plt.show()
```

* Histograma de Propinas: **`tip`**

```python
# Histograma con Matplotlib
import matplotlib.pyplot as plt

# Histograma de la variable 'tip'
tips['tip'].hist()
plt.title('Distribución de Propinas (tip)')
plt.xlabel('Propina (tip)')
plt.ylabel('Frecuencia')
plt.show()
```

* Histograma del Tamaño del Grupo: **`size`**

```python
# Histograma con Matplotlib
import matplotlib.pyplot as plt

# Histograma de la variable 'size'
tips['size'].hist()
plt.title('Distribución del Tamaño del Grupo (size)')
plt.xlabel('Tamaño del Grupo (size)')
plt.ylabel('Frecuencia')
plt.show()
```

##### Diagrama de Cajas: Box Plot

Un **Diagrama de Cajas (Box Plot)** es una representación visual que describe varias características importantes de un conjunto de datos, como la mediana, los cuartiles y los valores atípicos (outliers).

La función **`plt.boxplot()`** se utiliza para crear un diagrama de cajas con **`matplotlib.pyplot`**.


* Box Plot del Total de la Cuenta: **`total_bill`**

```python
# Box Plot con Matplotlib
import matplotlib.pyplot as plt

# Box Plot de la variable 'total_bill'
plt.boxplot(tips['total_bill'])
plt.title('Diagrama de Cajas del Total de la Cuenta (total_bill)')
plt.ylabel('Total de la Cuenta')
plt.show()
```

* Box Plot de las Propinas: **`tip`**

```python
# Box Plot con Matplotlib
import matplotlib.pyplot as plt

# Box Plot de la variable 'tip'
plt.boxplot(tips['tip'])
plt.title('Diagrama de Cajas de Propinas (tip)')
plt.ylabel('Propina')
plt.show()
```

##### Gráfico de Dispersión: Scatterplot

Un **Gráfico de Dispersión (Scatterplot)** es una representación visual que utiliza puntos para mostrar la relación entre dos variables numéricas.

* Scatterplot de **`total_bill`** y **`tip`**

```python
# Scatterplot con Matplotlib
import matplotlib.pyplot as plt

# Scatterplot de la variable 'total_bill' y 'tip'
plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Relación entre el Total de la Cuenta y la Propina')
plt.xlabel('Total de la Cuenta (total_bill)')
plt.ylabel('Propina (tip)')
plt.show()
```

## OBJETIVOS DEL TRABAJO
1.- Comprenda la importancia del preprocesamiento en el ciclo de análisis de datos.

2.- Practique técnicas de limpieza, integración, transformación, reducción y discretización.

3.- Desarrolle un flujo reproducible en su ordenador (sin necesidad de clústeres o nube).

### Aplicar un proceso completo de preprocesamiento de datos sobre un conjunto real, documentando las decisiones tomadas y mostrando los efectos de las transformaciones en la calidad de los datos.

# Ajustar la regresión lineal con el dataset

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size = 0.25, random_state = 0)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_train, y_train)
y_pred = lin_reg_2.predict(X_test)
print(lin_reg_2.score(X_train,y_train))
print(lin_reg_2.score(X_test,y_test))
```

# Visualización de los resultados del Modelo Lineal

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lin_reg_2.predict(X_poly), color = "blue")
plt.title("Modelo de Regresión Lineal")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```
